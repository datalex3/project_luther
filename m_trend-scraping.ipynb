```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import os

url = 'https://www.motortrend.com/cars/toyota/prius-v/2012/'
response = requests.get(url)
page = response.text
soup = BeautifulSoup(page,"lxml")
```

```python
string = soup.find(class_='mobile-scroller').find_all('span')
string
```

```python
car = soup.find_all(itemprop='itemListElement')
car
```

```python
string = soup.find(class_='mobile-scroller').find_all('span')
price = string[1].text
price = price.replace('\n','').replace('\t','').replace(',','')
price = int(price)
accel = string[3].text
accel = float(accel)
cityMPG = string[6].text
cityMPG = int(cityMPG)
hiMPG = string[8].text
hiMPG = int(hiMPG)
MPG = (cityMPG+hiMPG)/2
cargo = string[10].text
cargo = int(cargo)
car = soup.find_all(itemprop='itemListElement')
make = car[2].text
model = car[3].text
year = car[4].text
print('Make:  ',make)
print('Model: ',model)
print('Year:  '  ,year)
print('Price: ',price)
print('0-60mph: ',accel)
print('MPG:   ',int(MPG))
print('Cargo: ',cargo)
```
